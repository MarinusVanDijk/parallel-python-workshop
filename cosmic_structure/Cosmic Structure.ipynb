```python
import numpy as np
from matplotlib import pyplot as plt
from cosmic_structure import (Box, random_density, compute_potential, run_model, plot_structure)
```

```python
def power_spectrum(k):
    return np.exp(-k**2 / 100) / k**0.5

box = Box(N=256, L=50.0)

initial_density = random_density(box, power_spectrum)

fig, ax = plt.subplots(figsize=(8,8))
ax.pcolormesh(*box.grid_bounds, initial_density);
```

```python
potential = compute_potential(box, initial_density)

fig, ax = plt.subplots(figsize=(8,8))
ax.pcolormesh(*box.grid_bounds, potential);
```

```python
structure = run_model(box, potential, time=2.0)

fig, ax = plt.subplots(figsize=(8,8))
plot_structure(box, structure, xlim=[-20,20], ylim=[-20,20], ax=ax, point_scale=2.0, line_scale=2.0)
```

```python
fig, ax = plt.subplots(1, 3, figsize=(24,8))
for i, t in enumerate([1.0, 2.0, 4.0]):
    structure = run_model(box, potential, time=t)
    plot_structure(box, structure, xlim=[-20,20], ylim=[-20,20], ax=ax[i], point_scale=2.0, line_scale=2.0)
```
